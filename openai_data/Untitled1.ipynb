```python
def traverse_messages(mapping, current_id, messages_x):
    """
    Recursive function to traverse the message tree.
    """
    message_info = mapping.get(current_id, {})
    message = message_info.get('message')

    if message and message.get('content') and message['content']['parts']:
        author_role = message['author']['role']
        content = message['content']['parts'][0].strip()
        if content:  # Boş içerikleri atla
            if author_role == 'user':
                labeled_message = f"Soru: {content}"
            elif author_role == 'assistant':
                labeled_message = f"Cevap: {content}"
            else:
                labeled_message = f"{author_role.capitalize()}: {content}"
            messages_x.append(labeled_message)
    
    # Traverse children
    children = message_info.get('children', [])
    for child_id in children:
        traverse_messages(mapping, child_id, messages_x)

mapping = data.get('mapping', {})

# Mesajları etiketlemek için boş bir liste oluştur
messages_x = []

# Kök düğümü bulmak (parent = None)
root_ids = [msg_id for msg_id, msg in mapping.items() if msg.get('parent') is None]

# Her kök düğüm için traverse işlemini başlat
for root_id in root_ids:
    traverse_messages(mapping, root_id, messages_x)

# Etiketlenmiş mesajları yeni satırlarla birleştir
message = "\n\n".join(messages_x)
print(message)
```

```python
import json
import requests
from colorama import Fore, Style
import json
import pandas as pd
import ssl
import http.client



class ConversationProcessor:
    def __init__(self, conversation_id):
        """
        Initializes the ConversationProcessor with the given conversation ID and authorization token.

        :param conversation_id: The ID of the conversation to process.
        :param auth_token: Bearer token for authorization.
        :param host: The API host. Default is 'chatgpt.com'.
        """
        self.conversation_id = conversation_id
        self.headers = {
            
    'Accept': '*/*',
    'Accept-Language': 'tr-TR,tr;q=0.9,en-US;q=0.8,en;q=0.7',
    'Authorization': 'Bearer eyJhbGciOiJSUzI1NiIsInR5cCI6IkpXVCIsImtpZCI6Ik1UaEVOVUpHTkVNMVFURTRNMEZCTWpkQ05UZzVNRFUxUlRVd1FVSkRNRU13UmtGRVFrRXpSZyJ9.eyJzZXNzaW9uX2lkIjoiTE1lS01LME1MQmpCb3hpT2R6Z3FVUVpoY19jaEVfQ18iLCJodHRwczovL2FwaS5vcGVuYWkuY29tL3Byb2ZpbGUiOnsiZW1haWwiOiJ5aWxkaXJpcmFobWV0MjFAZ21haWwuY29tIiwiZW1haWxfdmVyaWZpZWQiOnRydWV9LCJodHRwczovL2FwaS5vcGVuYWkuY29tL2F1dGgiOnsicG9pZCI6Im9yZy1JYW9wZFFtZ1UySnRPYjBheE9wVkFqUTUiLCJ1c2VyX2lkIjoidXNlci1tNHFoeWFnVE9xMXk5dGlpU3BnWVM4REkifSwiaXNzIjoiaHR0cHM6Ly9hdXRoMC5vcGVuYWkuY29tLyIsInN1YiI6Imdvb2dsZS1vYXV0aDJ8MTA2NzQwNzQwNjIyNzA3NTEzNDY2IiwiYXVkIjpbImh0dHBzOi8vYXBpLm9wZW5haS5jb20vdjEiLCJodHRwczovL29wZW5haS5vcGVuYWkuYXV0aDBhcHAuY29tL3VzZXJpbmZvIl0sImlhdCI6MTczMTMwODc4NSwiZXhwIjoxNzMyMTcyNzg1LCJzY29wZSI6Im9wZW5pZCBlbWFpbCBwcm9maWxlIG1vZGVsLnJlYWQgbW9kZWwucmVxdWVzdCBvcmdhbml6YXRpb24ucmVhZCBvcmdhbml6YXRpb24ud3JpdGUgb2ZmbGluZV9hY2Nlc3MiLCJhenAiOiJUZEpJY2JlMTZXb1RIdE45NW55eXdoNUU0eU9vNkl0RyJ9.bEudyT1tM12nM68yaFgTKYuQ5id3iZ3oU6Q5CKwPzvMb3-1WsbEeAf8ds2rLiIhRsS3ioTCbbFvqrdNKnUH9JsG5kj2jba8AAm_QmerhfLEArLOcZbiVxEZ2GuavA7qei5VLchybcZpLvsB8Mw8GtEq5AmAcYR_wfHLOGc21wTRCUrHXTYiv3vUcxkOp3hmGxzxraFAUshGkq5quv5QBdNnY9qHvp67ZLXlYh9g5cXyCOoSFgjOmoLz48N8NMWNi7_bzsQcbCMjvRKffstdVB3-08wX88t-_4LgkfP6kGnSHTL92LG9kAYT7LVBOpYhrbpCCsbX0cjOCnKDBR3_tIQ',
    'Cache-Control': 'no-cache',
    'Connection': 'keep-alive',
    'Cookie': 'oai-did=b17603b7-ff55-46d8-911d-629c5dc08d05; oai-hlib=true; oai-nav-state=1; __Host-next-auth.csrf-token=2c0122b41e4fef60353287591d4173e6a09e386f27082254ef5997b17b638157%7Cd24c64205ae2e63e81f1076533d1dd2eeb040b6500d45fbb62b384c159ae4fb8; __Secure-next-auth.callback-url=https%3A%2F%2Fchat-identity-edge-onramp.unified-6.api.openai.com; _cfuvid=YRC8nnBGwzaVULTIt2aKdoUn1Bh7zeKRi_stAELCrME-1731475854596-0.0.1.1-604800000; oai-sidebar=open-pinned; __cflb=0H28vzvP5FJafnkHxj4GQsQCJmEZYHymSwRfCghuXvM; _uasid="Z0FBQUFBQm5ORWtGZklDSF9FdkJvUWZza0dETmVCRDlrZ0Q3VDVaekhLNmRPX29SYng2SzNyYWxuRUs2NnJfUHJ6MWlfcFRTUTNpZDhpNGZLM1I4UnV5QkhZNUpzSTdQT2Jza0tHSHpuY1Axa3h1XzB1eEJJWTFDZ09iMWtjVUc1UHc4ZFFLM3MzekMxczdERjY2ckduSnAzTnpaWE1rVmctV2VPMUs5SzVXTXU2NDdEeFo3N2hVdi1HTGVmSFcyQktIY3ZQNHg1dERDRmZWeTZ3eDZqa0pFaTNIcTZTUWVKaUpvbGlSZlNzUUFYWkQzTnEtWWdmYXNJOE9nd2hjQUdJOENBb1M1c3NOcVpxLWd0UGMyNHdrWldKNXQyTHN6U21yNlY1eWI5S1ZEVS1GTXVZYWJveE9PNjFsTGR4bGRKM0pJVmxCWmFYaXI3Uld3bWdHRHJRUWJTaVFuaUhra3pBPT0="; _umsid="Z0FBQUFBQm5ORWtGc3hod19Za2R0bmlrb3MybFozWXRsc2JmQnRxTE5MMi1WZmNMQUcwYVVBZXRIbEpzUUFMUzAweVpoUmlRd3FWQW45dXlrQTRnSTc5eV9aYnVuMUREVTVFUnJFUG9rUW9YanFId3NCeC1uNzhTSm9TRm9lWVBIS0dwQV8xWkJmNnZzMC1rb2JDckdhOVo3dlVKRmFwWE1sLVBrUUZ4bUFnQWVteHJUU3dXd21mNG9nOWlIVWFIb1JVeTRINXJFdnR2Y2xCQWlpV0lWV3BUMnU2R0xRV1JzbmE5ZW4yN2VRcnpjNGdWQndxNHNlRT0="; cf_clearance=iQe1M5w5Iwd4Ej4.cpvZUJzu8ooSzSPJnYyqpC0QiQ8-1731480107-1.2.1.1-KJJRzB__iUAQmmfDvJXFXZ4z.lT24nWtftjccZa18Qby5OHGhnjAFTOWaDRGRBlhgpCr8Ba0yei9KUrK2rYKe1vuJeoqWJlYmk7p5Z6uq6HKh9bS8XFT.FEaB1GDSiyvtowB49K4v_Xd2faJxwy7XX4BeC63U_4FSBIIpaC_hzfDe5DDeRNfH9Agq1dH8Sswz2YXn..XD4lrl3d.crJ_zo0DtVWJ2yg2.2QG_LoJ6VydDXTQPPP_xsVAJ64T1L.CVLZ699jJgz74WDAfcJ.iEF16nL7QtH.ftjJJVBNwuItdcw7RK7rqRxSQPnqwMkF6LUmRbFF01X7XsJuEMDoeeK0DFl_44O_9oVF7PXj2m_0QOyWqN3jSVIptpu4hpY.cQomJ24QcvnZl8onAYzk6CQ; __Secure-next-auth.session-token=eyJhbGciOiJkaXIiLCJlbmMiOiJBMjU2R0NNIn0..QcG9Mo8xlVlkfQ_n.GCKs3-TIntMNtSZLq-mhaTho_G_lsjnX_sLfSNructilpSG4gNxS8DD1Xt_p81-EjNP072UfYmxuOf1BTp6TsIcnjSoybpCgo9H389H_yWMp316kBnHvjkUUkgCyvmYj_GowWxxGaEz_-8fblSpKL8rcz7ETP5565hHduCAUidPTTWwcn0wCXr_O2uvpsMJAed5G0FeSZH6tsBN4vZIym_UhOgK80IyV4ukLWuhtzLsRNoovNkjb7KSJJ7Kc_aqbdr4D7jZnHItK78eAGm4oV7udxNEODdEQAwjpTD_KZnuGgc-TiDCdDLHU6WwUJcfqJbuvywhGC2qCm7uO4YcwzGzILQNRzPQ3qdWvl8SKmWbWb3euwLnZvJ0i6cUwzT6Qss2sS08q578Nl8YWYXtp_Xs06S7RH--deDMQU_YgOaJiHN1-xNmLrdpWjWb1gjVIfyzJvzlY5KWk01ikPTyYb9QTuOYOcphKUkU3SadKU7EqJfesoSmcwSPtpcBHhEcGIrAYdkzgmqBPLwGKuSdzMGnkj66f_E5nW0pB6drihBfk3pYwHQj8oyWmOKG0yFaZdbdEnNGTBWxgHBwTckyxa00ne9jamj0oUtqza5J3wl9nTMU-_1SSOpKMu6ShohnBMdNFLWl21Nzn_8AvBRpb2U081ALl3z6Xwvq80kjRkKAa8Ma7xmNyEyzUtWnu1tKptgPOQM_vs9ZOMlnrA_vTUqXI4sYoZb7uwH-soA-da8H1VRuKx7nWdxFyK4fPaE_EKKfj8PpId-4fH1oaFMgf6uJxkOD1BdLU3Vv6vMVY08MquRW_CmFXk7MN7auKZZ0eECsq0YDVU6Vqr4oV4_9NLY3pghzsoUUIMrfgP-yumOQdJvELtINoNP4GI9SMmuf_Ao0PIPv7fV2FbltBp9k7kH0Om3dvofsWzO9YfU3JU4t6a0V88mJm3h5aIKPqew1c52l-r3ElL7kQCOqRyBy-zfBWhN9w5g473tL8DzP8S2FrBqitwg1ifeuwy-PcctXBTeRBF1EwdjKN3-81iW0C8sW5xxkJOoDFFZQez-XTTLND8njjd4kpRVgCowj_Y46cQAOKv90oR0OYcae0OdSKUP7zv7zk8lXbqaRwEtwW-ONeWf3sSpD_3ndf027dJDKSc8YpC6LJvFI9f9m9-FzvhKdkVWonzeYVoBFH7JJdxALIdN9fBTGp3w8taVUf-xhxNUr6aB6JCv_FHMf1rkzt1pFwVZnK6H-LyzHLOItAvhaOsDPQY87EwWfnyCdnAutngEqDhgedvspD4IlC2brHio3Ns7Yq5srmXik4WXd3Z3yfSLuHBAjtwJOPsaWVzaQ0sFWcoMSSOJow5HNpcL5AG-3RB03omGyiXzo6YaoDp6upfVMGV5IIrsSsF_uBbBJLvs2eXiXZSDwYwke1T3qRewIARXKl4R3GLPTRATKPNXn0cPR-HXCdTYmJF34GjOKxFi35OepaQ6GeFBFOIhbOLt7fdTuHfZfMpQXlfOC0_HNcDE72ieOL4_kIlEA_IxIE18slxiOvuLdIHI3Qmju53hrL6GLZlKqqPAWybLVIIsaPGvF6oeuVDUvUwSpBWOl_0J6hJ37ENQ-_0JuJkWr6G09I0QJw1UjY9oUiXPNhHGFwHkjbP1C1d6XZMri2KGCX0h9B2kPGMlSPHay5TnVULwsTTWsi7ehewnNmz1Ze_lN-it5lDVfHmrgdWKv7CRqhMS21tkUJVqaw-Pq9-yNG6cKZ7kapga2VxyTLlMdM-4qfTZ3fcb0KpR4Q_UqcWXzFxkYhUuxXJ4Xl8Q37uJCr4sG2L6P8wViV-ZhUPiQJbh9dzPHUyLvaVqKqkfUXQfaQl5z9k8TCkCh1wGEErrFM1F0wnuGsW9UNosFZ5qiAtO8bLkaqZQlkfTcPzqlpQ-fUBCyeMCnlgcNTU2OcukihA4wfu7Phoe5fxzOmg1ZNSSEszjc-yQZpSrG-JVWSzxZQWrGaqDph0a8mavCJwBrmHfSPfuOmL061dwxd-fGBp5v5moG_2JvkLKt8EqDSdF6y3SokPvIdmHa3OsB6dc53wbO6qSbaLgbgGhDm_Nj9Lain-aNUeuKA-Hecpw5DYz_C_hrVJIH6_9qEGhvgmc4GztUwlQmC52IsRKyRylFDibzCKAB8XBu1B_WSMm-P28QLVm35VWnBUUrzzpgW_8wXkslPCTPFoQ-Uw1aM_o83trIIzsCrdakRatWyn7vm0QjlzBQF7GHOlLTHVgL6rLsC5abn5SjFRr-Zw3plKWwG05_8DkzURvhdTnPXQMmp3u52nAkHOT4AqYwnNRDsuNG36BjEhk5jSAPKYsoEQmsycuP0m0PngCBAJ_lkzponX5qySAo2K2TnIt4C8Q2vO0bdW_PDzjg6kL94Exe4Bzua0er74-32HKsMmw_nHj-7_4yE-AL20es3PAx85ZwHnsC_OjXV8rMH-l5ZH52Luq2UnsVfcxFjzb1NC5W3eS5o_FP_AtULIh3CQ9r4IGNWb5S_E3hxCDi6d47OBOwie6p_zc0BXgEgKrwF2qRVsGiYTi-fHuAah46zPenzdp5n52EEVcZvlnxFtOF9GBTWCOf50zvw_-v8UZq-ejw1jB2hWG7pTy_K7b2rmJRuJGQ-HYfPUE8_2lOWd9hs_ASApT4RWTnATZGVvouYS69T72N-gNuiEuoyj0rKH_ZNctbtF7dG_enQ69KNuyLNsbApKLgQW9OaW0b97Wkg_B0-x-f8GVSimvENgBuXBQOB5_L1xobGdI30PLxai-b6GUl3zb1o1fxHHKlt3a6-0yelwmy8m-tj4jIGcR61G5dDm2kdjGaLo-ZKFtXPSPKNP2Ym7CJcUdQ66kHlOpHERiGnMlOEsPUuTvdPkyTHXReopVrWfqzbrnNQT9dk7egDT5PydW277K3qRNxuAerLW1IIhn3quZI2.qL_NgmirtDmjS8COj6NNww; __cf_bm=4fCqA8Va7awWLtf3nRn_4ay5mgopDOJikeyFqW4KkKg-1731480122-1.0.1.1-Lz1_CUp7ty1X9CcJe6c541HrY4Qvrl0RSsCVEqAfDYAYDpi.t26tc54N4HxYwDDlTXG0Iy.eueE2Lmo8z2mIsQ; _puid=user-m4qhyagTOq1y9tiiSpgYS8DI:1731480124-XRvwrIcRXt1E6x8ezihjNDU%2BLaou59On3QJ016ZBpec%3D; oai-sc=0gAAAAABnNEo9izcp5faczpZsq8i53egp92P8u2dwDWmCK9OE0xWJpJo_bI6249ZHAwG7Yf11ST2JI7Co9y73BEscKSCcWGUpaWhvhvOtJDpjoWMtmbZNVhdCpKr5k4GekRpikGQEX2vXwHt4FVLkVR8AISCTVPXIg1EixiOaCpM-pAJBvRQzrWE8cEx5iRjmuokeLC1lNgvF8qqykpouAT6YZUp8dYAcoQ; _dd_s=rum=0&expire=1731481106541&logs=1&id=4a08f7bf-0551-4b38-8a71-643eb6855ce0&created=1731479219487',
    'OAI-Device-Id': 'b17603b7-ff55-46d8-911d-629c5dc08d05',
    'OAI-Language': 'tr-TR',
    'Pragma': 'no-cache',
    'Referer': 'https://chatgpt.com/g/g-z77yDe7Vu-books/c/672deee4-f3b8-800e-96f8-928547930e0f',
    'Sec-Fetch-Dest': 'empty',
    'Sec-Fetch-Mode': 'cors',
    'Sec-Fetch-Site': 'same-origin',
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/129.0.0.0 Safari/537.36',
    'sec-ch-ua': '"Google Chrome";v="129", "Not=A?Brand";v="8", "Chromium";v="129"',
    'sec-ch-ua-mobile': '?0',
    'sec-ch-ua-platform': '"Windows"',
}
 
        
    def fetch_conversation_data(self):


        try:
            context = ssl._create_unverified_context()
            conn = http.client.HTTPSConnection('chatgpt.com',context=context)
            conn.request(
                'GET',
                '/backend-api/conversation/'+str(self.conversation_id),
                headers=self.headers
            )
            response = conn.getresponse()
            
            data = response.read()
            data = json.loads(data)
           
            return data
            
        except requests.exceptions.HTTPError as http_err:
            print(f'HTTP error occurred: {http_err}')
            
        except Exception as err:
            print(f'Other error occurred: {err}')
            return None

    def traverse_messages(self, mapping, current_id, messages_x):
        """
        Recursive function to traverse the message tree and label messages.

        :param mapping: The mapping dictionary from the conversation data.
        :param current_id: The current message ID being traversed.
        :param messages_x: The list to append labeled messages.
        """
        message_info = mapping.get(current_id, {})
        message = message_info.get('message')

        if message and message.get('content'):
            try:
                author_role = message['author']['role']
                content = message['content']['parts'][0].strip()
                if content:  # Boş içerikleri atla
                    if author_role == 'user':
                        # "Soru:" etiketini daha kalın hale getirmek için üç yıldız kullandık
                        labeled_message = f"***Soru:*** {content}"
                    elif author_role == 'assistant':
                        # "Cevap:" etiketini kalın yapmak için çift yıldız kullandık
                        labeled_message = f"**Cevap:** {content}"
                    else:
                        labeled_message = f"{author_role.capitalize()}: {content}"
                    messages_x.append(labeled_message)
            except:
                pass

        
        children = message_info.get('children', [])
        for child_id in children:
            self.traverse_messages(mapping, child_id, messages_x)

    def get_labeled_messages(self):
        """
        Retrieves and processes the conversation data to return labeled messages.

        :return: A string containing all labeled messages separated by two newlines.
        """
        data = self.fetch_conversation_data()
        if not data:
            return "Veri alınamadı."

        mapping = data.get('mapping', {})
        if not mapping:
            return "Mesajlar bulunamadı."

        # Mesajları etiketlemek için boş bir liste oluştur
        messages_x = []

        # Kök düğümü bulmak (parent = None)
        root_ids = [msg_id for msg_id, msg in mapping.items() if msg.get('parent') is None]

        # Her kök düğüm için traverse işlemini başlat
        for root_id in root_ids:
            self.traverse_messages(mapping, root_id, messages_x)

        # Etiketlenmiş mesajları yeni satırlarla birleştir
        message = "\n\n".join(messages_x)
        return message

  


conversation_id = "672ca105-43c8-800e-a22c-fcdabf0fdd25"
processor = ConversationProcessor(conversation_id)
labeled_messages = processor.get_labeled_messages()
print(labeled_messages)
```
